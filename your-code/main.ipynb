# Challenge 3 - Constructing Confidence Intervals

We will test whether the hourly wage of all hourly workers is significantly different from $30/hr.

In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. Is $30/hr within that interval?

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
#For the calculation of 95% confidence interval, we need
#1 The mean
#2 The number of observation
#the variance or standard deviation of the distribution

mean = salaries_hourly["Hourly Rate"].mean()
standard_deviation = salaries_hourly["Hourly Rate"].std()
n = len(salaries_hourly)
confidence_level = 0.95
standard_error = standard_deviation/np.sqrt(n)
degree_of_freedom = n - 1
```

```python
#Method 1.
confidence_interval = t.interval(confidence_level, degree_of_freedom, loc = mean, scale = standard_error)
confidence_interval
#the $30/hr is not within the confidence interval
```

```python
#Alternative 2
print("The mean:", mean)
print("The standard deviation:", standard_deviation)
print("The number of samples:", n)
print("The upper boundary: ", mean + 2* (standard_deviation/np.sqrt(n)))
print("The lower boundary: ", mean - 2* (standard_deviation/np.sqrt(n)))
```

This is fine if we have thousands of worker data. But what if we have only 100 workers data?

Sample 100 workers and re-construct the 95% confidence interval. Is the interval wider of narrower? And why?
Do you still encapsulate the $30/hr mark in this case?

```python
#sample 100 workers
salaries_hourly_100 = salaries_hourly.sample(n = 100, random_state = 1)
```

```python
#computing the confidence interval for 100 sample
mean_100 = salaries_hourly_100["Hourly Rate"].mean()
standard_deviation_100 = salaries_hourly_100["Hourly Rate"].std()
n_100 = len(salaries_hourly_100)
confidence_level = 0.95
standard_error_100 = standard_deviation_100/np.sqrt(n_100)
degree_of_freedom = n_100 - 1
confidence_interval = t.interval(confidence_level, degree_of_freedom, loc = mean_100, scale = standard_error_100)
confidence_interval
```
